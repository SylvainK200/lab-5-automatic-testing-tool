```python
import pickle
import matplotlib.pyplot as plt
```

```python
number_to_words = []
with open("results_number_to_words.pkl", 'rb') as f:
    number_to_words = pickle.load(f)
```

```python
print(number_to_words[0])
```

```python
strong_password = []
with open("results_strong_password.pkl", 'rb') as f:
    strong_password = pickle.load(f)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example dataset: Three separate lists of values
data1 = [res['generator']['coverage']['percent_covered'] for res in number_to_words ]
data2 = [res['llm']['coverage']['percent_covered'] for res in number_to_words]

improvement = [res['improvement'] for res in number_to_words]

# X positions for each dataset
x1 = np.arange(len(data1))
x2 = x1 + 0.2  # Slight offset to avoid overlap


# Create the plot
fig, ax = plt.subplots(figsize=(8, 5), nrows=1, ncols=2)

# Plot the bars for each dataset
ax[0].bar(x1, data1, width=0.2, color='blue', label='Test generator')
ax[0].bar(x2, data2, width=0.2, color='red', label='Initial test')

ax[1].bar(x2, improvement, width=0.2, color='gray', label='Improvement')
# Add labels and title
ax[0].set_xlabel('Runs')
ax[0].set_ylabel('coverage')
ax[0].set_title('Coverage plot comparison ')
ax[0].set_xticks(x1 + 0.2)  # Set x-ticks to be in the center of the grouped bars
ax[0].set_xticklabels([str(i) for i in range(len(data1))])  # Label the x-axis with the indices
ax[0].legend()


ax[1].set_xlabel('Runs')
ax[1].set_ylabel('improvement')
ax[1].set_title('Improvement plot ')
ax[1].set_xticks(x1 + 0.2)  # Set x-ticks to be in the center of the grouped bars
ax[1].set_xticklabels([str(i) for i in range(len(data1))])  # Label the x-axis with the indices
ax[1].legend()

# Show the plot
plt.show()
```
